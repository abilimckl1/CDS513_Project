#### One Hot Encoding

```python
data['belongs_to_collection'] = data['belongs_to_collection'].astype('category')
data['belongs_to_collection'] = data['belongs_to_collection'].cat.codes

data['original_language'] = data['original_language'].astype('category')
data['original_language'] = data['original_language'].cat.codes

data['production_companies'] = data['production_companies'].astype('category')
data['production_companies'] = data['production_companies'].cat.codes

data['spoken_languages'] = data['spoken_languages'].astype('category')
data['spoken_languages'] = data['spoken_languages'].cat.codes

data['genres'] = data['genres'].astype('category')
data['genres'] = data['genres'].cat.codes

data['production_countries'] = data['production_countries'].astype('category')
data['production_countries'] = data['production_countries'].cat.codes

data['adult'] = data['adult'].astype('category')
data['adult'] = data['adult'].cat.codes
```

#### Drop unused features

```python
df_new = data.drop(["title","release_date","overview"], axis = 1)
df_new
```

#### Training and Testing the Model

#### Splitting data into training and test datasets

```python
features = ['adult', 'belongs_to_collection','genres','runtime','original_language','production_companies','production_countries','spoken_languages','budget','vote_average','vote_count','popularity']                # create a Python list of feature names
target = ['revenue']                                     # Define the target variable
```

```python
df_new.head()
```

```python
df_new[features]
```

```python
data[target]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_new[features], df_new[target], test_size=0.05, random_state=5000)
```

```python
print('Train cases as below')
print('X_train shape: ',X_train.shape)
print('y_train shape: ',y_train.shape)
print('\nTest cases as below')
print('X_test shape: ',X_test.shape)
print('y_test shape: ',y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test.head()
```

```python
y_test.head()
```

#### Fitting Linear Regression Model

```python
#Instantiating the model
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression(fit_intercept=True)
```

```python
lr_model.fit(X_train, y_train)
```

#### Interpreting Model Coefficients

```python
print('Intercept:',lr_model.intercept_)          # print the intercept 
print('Coefficients:',lr_model.coef_)  
```

```python
X_train.columns
```

```python
(lr_model.coef_).T
```

```python
pd.DataFrame((lr_model.coef_).T,index=X_train.columns,\
             columns=['Co-efficients']).sort_values('Co-efficients',ascending=False)
```

##### y = 19 883 107.52 + 1.03e+06 * popularity + 3.99e+05 *original_language + 7.25e+04 * vote_count + 5.72e+04 * production_countries + 1.98e+00 * budget - 4.38e+03 * belongs_to_collection - 8.58e+03 * spoken_language - 1.32e+04 * genres - 1.16e+05 * runtime - 3.98e+06 * vote_average

#### Using the Model for Prediction

```python
y_pred_train = lr_model.predict(X_train)  
```

```python
y_pred_train                                                         # make predictions on the training set
```

```python
y_pred_test = lr_model.predict(X_test)                                  # make predictions on the testing set
```

```python
y_pred_test
```

#### Model evaluation 

#### Model Evaluation using __metrics.__

__Mean Absolute Error__ (MAE)

```python
MAE_train = metrics.mean_absolute_error(y_train, y_pred_train)
MAE_test = metrics.mean_absolute_error(y_test, y_pred_test)
```

```python
print('MAE for training set is {}'.format(MAE_train))
print('MAE for test set is {}'.format(MAE_test))
```

__Mean Squared Error__ (MSE)

```python
MSE_train = metrics.mean_squared_error(y_train, y_pred_train)
MSE_test = metrics.mean_squared_error(y_test, y_pred_test)
```

```python
print('MSE for training set is {}'.format(MSE_train))
print('MSE for test set is {}'.format(MSE_test))
```

__Root Mean Squared Error__ (RMSE)

```python
RMSE_train = np.sqrt( metrics.mean_squared_error(y_train, y_pred_train))
RMSE_test = np.sqrt(metrics.mean_squared_error(y_test, y_pred_test))
```

```python
print('RMSE for training set is {}'.format(RMSE_train))
print('RMSE for test set is {}'.format(RMSE_test))
```

```python
RMSE_test/df_new['revenue'].mean()
```
