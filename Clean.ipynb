# Data Cleaning

```python
import json
import ast
import numpy as np
```

```python
# Function to extract names from JSON
def stringify_extract(json_str):
    if pd.isna(json_str):
        return np.nan
    try:
        # Use ast.literal_eval to safely evaluate the string representation of a Python object
        data = ast.literal_eval(json_str)
        if isinstance(data, list):
            # Extract the "name" values from a list of dictionaries
            names = [item['name'] for item in data if 'name' in item]
            # Join the names into a single string
            names_str = ', '.join(names)
        elif isinstance(data, dict):
            # Extract the "name" value from a single dictionary
            names_str = data.get('name', "")
        else:
            names_str = ""
    except (ValueError, SyntaxError, TypeError) as e:
        print(f"Error decoding JSON: {e}")
        names_str = ""

    # Check if names_str is empty, if yes return NaN
    if not names_str:
        return np.nan

    return names_str
```

## keywords

```python
df_keywords = pd.read_csv(input+"\\keywords.csv")

# List of columns to apply the stringify_extract function
columns_to_extract = ['keywords']
# Apply the stringify_extract function to each specified column
for column in columns_to_extract:
    df_keywords[column] = df_keywords[column].apply(stringify_extract)
```

## Extracting keywords based on overview

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from nltk.corpus import names

# Load the stopwords and names
stopwords_list = list(ENGLISH_STOP_WORDS)

import nltk
nltk.download('names')

# Get lists of male and female names from the nltk corpus
male_names = names.words('male.txt')
female_names = names.words('female.txt')

# Combine male and female names into a single set of names (case insensitive)
all_names = set(name.lower() for name in male_names + female_names)

# Combine stopwords and names into a single set for filtering
filter_words = set(stopwords_list).union(all_names)

def extract_keywords(text, filter_words):
    if pd.isna(text):
        return None
    words = text.replace(',', '').split()
    filtered_words = [word.lower() for word in words if word.lower() not in filter_words]
    return ', '.join(filtered_words) if filtered_words else None
```

```python
for index, row in df_movies.iterrows():
    if pd.isna(row['keywords']) and not pd.isna(row['overview']):
        df_movies.at[index, 'keywords'] = extract_keywords(row['overview'], stopwords_list)
```
