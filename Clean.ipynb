# Data Cleaning

```python
import json
import ast
import numpy as np
```

```python
# Function to extract names from JSON
def stringify_extract(json_str):
    if pd.isna(json_str):
        return np.nan
    try:
        # Use ast.literal_eval to safely evaluate the string representation of a Python object
        data = ast.literal_eval(json_str)
        if isinstance(data, list):
            # Extract the "name" values from a list of dictionaries
            names = [item['name'] for item in data if 'name' in item]
            # Join the names into a single string
            names_str = ', '.join(names)
        elif isinstance(data, dict):
            # Extract the "name" value from a single dictionary
            names_str = data.get('name', "")
        else:
            names_str = ""
    except (ValueError, SyntaxError, TypeError) as e:
        print(f"Error decoding JSON: {e}")
        names_str = ""

    # Check if names_str is empty, if yes return NaN
    if not names_str:
        return np.nan

    return names_str
```

## keywords

```python
df_keywords = pd.read_csv(input+"\\keywords.csv")

# List of columns to apply the stringify_extract function
columns_to_extract = ['keywords']
# Apply the stringify_extract function to each specified column
for column in columns_to_extract:
    df_keywords[column] = df_keywords[column].apply(stringify_extract)
```

```python
df_movies = pd.read_csv(input+"\\movies_metadata.csv")
df_movies = df_movies[['id','title','release_date','adult','belongs_to_collection',\
                       'genres','runtime','original_language','overview',\
                       'production_companies','production_countries','spoken_languages',\
                       'revenue','budget','vote_average','vote_count','popularity']]
```
